```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import scienceplots

# ---- match plot style ----
plt.style.use(["science", "grid"])

TITLE_SIZE = 30
AXIS_SIZE = 26
TICK_SIZE = 24
LEGEND_SIZE = 22
LINEWIDTH = 3

mpl.rcParams.update({
    "font.size": TICK_SIZE,
    "axes.titlesize": TITLE_SIZE,
    "axes.labelsize": AXIS_SIZE,
    "xtick.labelsize": TICK_SIZE,
    "ytick.labelsize": TICK_SIZE,
    "legend.fontsize": LEGEND_SIZE,
})

# ---- colors ----
GCACI_COLOR_MAP = mpl.cm.Reds
OTHER_COLOR = "#0700C8"
MARGINAL_COLOR = "#2CA02C"

# ---- legend handles ----
handles = [
    Line2D([0], [0], color=MARGINAL_COLOR, marker='o', linestyle='-',
           linewidth=LINEWIDTH + 2, markersize=10),
    Line2D([0], [0], color=OTHER_COLOR, marker='o', linestyle='-',
           linewidth=LINEWIDTH + 2, markersize=10),
    Line2D([0], [0], color=GCACI_COLOR_MAP(0.65), marker='o', linestyle='-',
           linewidth=LINEWIDTH + 2, markersize=10),
]

labels = [
    r"UP-OCP",
    r"POGO",
    r"GCACI ($\eta=1.0$)",
]

# ---- legend-only figure ----
fig, ax = plt.subplots(figsize=(12, 1))
ax.axis("off")

ax.legend(
    handles,
    labels,
    loc="center",
    frameon=False,
    ncol=3,
    handlelength=2.0,
    columnspacing=2.0,
    fontsize=LEGEND_SIZE,
)

fig.savefig("legend.pdf", bbox_inches="tight")
```
